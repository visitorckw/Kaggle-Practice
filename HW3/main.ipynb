````python
# 載入必要套件

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error as MSE
import numpy as np
import pandas as pd
````

## 資料前處理

## 觀察資料的特性

### TODO #3: 利用 Seaborn 畫箱型圖，並進行說明

### TODO #4: 利用 Seaborn 進行 Heatmap 作圖，並進行說明
- 觀察每一個特徵與房價 (MEDV) 的關聯程度
- 使用 dr.corr() 來計算 feature 和 feature 之間的關係係數
- 數值愈高表示愈正相關，數值愈低表示愈負相關

````python
plt.figure(figsize=(16, 9))
# Please use a function of Pandas here.
print(df.corr())
# Plot a heatmap with Seaborn
sns.heatmap(df.corr(),annot = True) # 塗顏色
plt.show()
````

- 可以觀察到 RM, PTRATIO, LSTAT 三個特徵與MEDV相關係數超過0.5
- 所有特徵以 CHAS和MEDV的相關係數最低

### TODO #5: 挑選特徵，並進行說明
- 觀察 heatmap 中具有"顯著"正相關或是"顯著"負相關的特徵
- 把這些特徵進行 concatenation，得到新的 dataframe

````python
X = pd.DataFrame(np.c_[ df['RM'], df['PTRATIO'], df['LSTAT'] ], # Please put features inside the bracket
                 columns = ['RM', 'PTRATIO', 'LSTAT'] # Please put feature names here 
                )
````

- 在挑選特徵時 只挑選相關係數超過0.5以上的特徵當作feature
- 總共選了三組feature

### TODO #6: 呈現特徵分佈，並進行說明

## 訓練模型前的準備
### TODO #7: 切分資料
- 利用 train_test_split，將資料切成 train set 和 test set
- 設定 train set 的比例為全部資料的 80%
- 請記得設定 random_state 以控制隨機過程

````python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, dataset.target, test_size = 0.2, random_state = 1) 

print(X_train.shape) #(404, 2)
print(X_test.shape) #(102, 2)
print(Y_train.shape) #(404, )
print(Y_test.shape) #(102, )
````

## 訓練模型並進行評估

### TODO #8: 訓練 Linear Regression 模型，並說明訓練結果
- 利用 scikit-learn，將資料 fit 在 LinearRegression
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

````python
# Modeling
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,Y_train)

````

````python
# plotting the y_test vs y_pred
Y_pred = reg.predict(X_test)
plt.scatter(Y_pred, Y_test)
plt.xlabel('Y_pred')
plt.ylabel('Y_test')
plt.show()
print('R2: ', reg.score(X_test, Y_test))
print('MSE: ', MSE(Y_test, Y_pred))



````

- 根據輸出的數據可以看到 R2約為0.7左右 而MSE為30左右
- 根據圖形觀察 數據分布隱約能看出有限性的關係 但仍然十分分散
- 訓練效果仍然不甚理想

### TODO #9: 訓練 Random Forest Regressor 模型
- 利用 scikit-learn，將資料 fit 在 RandomForestRegressor 模型
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

````python
from sklearn.ensemble import RandomForestRegressor
````

````python
# Modeling
reg = RandomForestRegressor(random_state = 1)
reg.fit(X_train,Y_train)


````

````python
# plotting the y_test vs y_pred
Y_pred = reg.predict(X_test)
plt.scatter(Y_pred, Y_test)
plt.xlabel('Y_pred')
plt.ylabel('Y_test')
plt.show()
print('R2: ', reg.score(X_test, Y_test))
print('MSE: ', MSE(Y_test, Y_pred))


````

- 根據輸出的數據可以看到 R2約為0.85左右 而MSE為14左右
- 根據圖形觀察 數據分布隱約能看出有限性的關係 比起線性回歸更符合憲性關係
- 訓練效果較為理想

### TODO #10: 參數調整，並進行說明
- 請嘗試使用 Grid search 來進行調參
    - 可以自己寫 for-loop
    - 或是使用 sklearn.model_selection 中的 GridSearchCV 函式

````python
from sklearn.model_selection import GridSearchCV
````

````python
# LinearRegression
param_grid = {'copy_X': (False, True), 'fit_intercept': (False, True), 'normalize': (False, True)}
grid = GridSearchCV(LinearRegression(), param_grid)
reg.fit(X_train,Y_train)
print('R2: ', reg.score(X_test, Y_test))
Y_pred = reg.predict(X_test)
print('MSE: ', MSE(Y_test, Y_pred))
````

- 將所有線性回歸模型可以調的參數都嘗試過一遍
- 訓練效果有了顯著的提升

````python
param_grid = {'bootstrap': (False, True), 'max_depth':(None, 10, 50, 100, 200, 500), 'n_estimators': (100, 50, 200), 
             'min_samples_split': (2, 5, 10), 'min_samples_leaf': (1, 5, 10, 50), 'oob_score0': (False, True),
              'warm_start': (False, True)}
grid = GridSearchCV(RandomForestRegressor(random_state = 1), param_grid)
reg.fit(X_train,Y_train)
print('R2: ', reg.score(X_test, Y_test))
Y_pred = reg.predict(X_test)
print('MSE: ', MSE(Y_test, Y_pred))
````

- 將多組參數都嘗試了許多種不同的可能
- R2與MSE仍與預設的參數訓練出來的model相同 說明預設的參數就已經是目前找到最佳的參數
- 訓練結果沒有明顯的提升
